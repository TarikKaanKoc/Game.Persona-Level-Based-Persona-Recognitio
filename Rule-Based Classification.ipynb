```python
# Ülkelere göre satışlardan toplam ne kadar kazanılmış?
df.groupby(['COUNTRY'])[['PRICE']].agg('sum')
```

```python
# Source  Türlerine göre satış sayıları nedir?
df.groupby(['SOURCE'])[['PRICE']].agg('count')
```

```python
# Ülkelere göre PRICE Ortalamaları nedir?
df.groupby(['COUNTRY'])[['PRICE']].agg('mean')
```

```python
# Source'lara göre price ortalamaları nedir?
df.groupby(['SOURCE'])[['PRICE']].agg('mean')
```

```python
# Country-Source kırılımında price ortalamaları nedir? (1.yöntem)
df.groupby(['COUNTRY','SOURCE'])['PRICE'].mean().unstack()
```

```python
# Country-Source kırılımında price ortalamaları nedir? (2. yöntem)
df.groupby(['SOURCE','COUNTRY'])[['PRICE']].agg('mean')
```

```python
# COUNTRY, SOURCE, SEX, AGE kırılımında ortalama kazançlar nedir?
df.groupby(['COUNTRY','SOURCE','SEX','AGE'])[['PRICE']].agg('mean')
```

```python
# Çıktıyı PRICE’a göre sıralamak ve 
```

```python
agg_df = df.groupby([x for x in df.columns[df.columns != 'PRICE']]).mean()
```

```python
# önceki işlemdeki çıktıyı daha iyi görebilmek için sort_values metodunu azalan olacak şekilde PRICE’a göre ayarlamak.
agg_df.sort_values(
    by = ['PRICE'],ascending = False, inplace = True)
```

```python
agg_df.head()
```

```python
# Rest Index
```

```python
agg_df.reset_index(inplace = True)
```

```python
agg_df.head()
```

```python
# Age değişkenini kategorik değişkene göre çevirmek ve agg_df’e eklemek.
```

```python
# • Aralıkları ikna edici şekilde oluşturduk:
```

```python
# • ‘0_18', ‘19_23', '24_30', '31_40', '41_70'
```

```python
agg_df['AGE_CAT']  = pd.cut(x = agg_df['AGE'], bins = [0,18,23,30,40,70], 
                        labels=['0_18', '19_23', '24_30', '31_40', '41_70'])
```

```python
agg_df.head()
```

```python
# Yeni seviye tabanlı müşterileri (persona) tanımlamak.
```

```python
agg_df['CUSTOMERS_LEVEL_BASED'] = ['_'.join(i).upper() for i in agg_df.drop(['AGE', 'PRICE'], axis=1).values]
                                   
agg_df
```

```python
# dataframe tekilleştirmek 
agg_df = agg_df.groupby(['CUSTOMERS_LEVEL_BASED'])[['PRICE']].agg('mean')
agg_df
```

```python
# Yeni müşterileri (personaları) segmentlere göre ayırmak. (Toplamda: 4 segmente ayırdım [A , B, C ,D])
# •Segmentleri CUSTOMER_SEGMENT isimlendirmesi ile değişken olarak agg_df’ Ekledik.
```

```python
agg_df['CUSTOMER_SEGMENT'] = pd.qcut(agg_df['PRICE'], 4, labels=['D','C','B','A'])
agg_df
```

```python
agg_df.sort_values(
    by = ['PRICE'],ascending = True, inplace = True)
agg_df.head()
```

```python
agg_df.reset_index(inplace = True)
```

```python
agg_df.head(50)
```

```python
# •Segmentlerin istatiksel betimlenmesi: (Segmentlere göre groupby yapıp price [mean, max, sum] istatistikleri gözlemlemek
agg_df.groupby(['CUSTOMER_SEGMENT'])[['PRICE']].agg(['mean','max','sum'])
```

```python
# Yeni gelen müşterileri sınıflandırıp, ne kadar gelir getirebileceklerini  tahmin etmek.
```

```python
agg_df[agg_df['CUSTOMERS_LEVEL_BASED'] == new_customer_2]
```
